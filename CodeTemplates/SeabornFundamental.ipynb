```python
race_counts = df['race'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Race Categories')

plt.axis('equal')  

plt.show()
```
