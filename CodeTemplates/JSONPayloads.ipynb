```python
response = requests.get(url, headers=headers)
data = response.json()

# Extracting top 5 movies from the response
top_movies = data['results'][:5]

# Extracting movie titles and ratings
movie_titles = [movie['title'] for movie in top_movies]
ratings = [movie['vote_average'] for movie in top_movies]

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.barh(movie_titles, ratings, color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Movie Title')
plt.title('Top 5 Movies on Chart')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest rating on top
plt.show()
```

```python
response = requests.get(url, headers=headers)
data = response.json()

# Extracting top 5 movies from the response
top_movies = data['results'][:5]

# Extracting movie titles and ratings, and multiplying ratings by a factor (e.g., 10) for a bigger change
movie_titles = [movie['title'] for movie in top_movies]
ratings = [movie['vote_average'] * 10 for movie in top_movies]

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.barh(movie_titles, ratings, color='skyblue')
plt.xlabel('Average Rating (Scaled)')
plt.ylabel('Movie Title')
plt.title('Top 5 Movies on Chart (Scaled Ratings)')
plt.gca().invert_yaxis()
plt.show()
```
